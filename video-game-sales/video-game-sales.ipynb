```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = pd.read_csv("vgsales.csv")
data = dt.copy()
data
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
data = data.reset_index(drop = True)
```

```python
data
```

```python
data["Year"] = data["Year"].astype(int)
```

```python
data.head(5)
```

```python
data = data.drop(["Rank"], axis = 1)
```

```python
data.head(5)
```

# En Çok Global Satış Yapan ilk 15 Oyun

```python
games_global_sales = data.head(15).groupby("Name")["Global_Sales"].sum().sort_values(ascending = False)
games_global_sales
```

```python
f, ax = plt.subplots(figsize = (10,6))
games_global_sales.plot(kind = "bar")
plt.xlabel("Oyun", fontfamily = "serif", size = 12)
plt.ylabel("Milyon $ Satış", fontfamily = "serif", size = 12)
plt.title("En Çok Global Satış Yapan ilk 15 Oyun", fontfamily = "serif", size = 12)
plt.show()
```

# Action türünde oyun çıkaran Publisher'lar - İlk 10

```python
action_df = data[data["Genre"] == "Action"]
publisher_counts = action_df["Publisher"].value_counts()
```

```python
f, ax = plt.subplots(figsize = (8,4))
publisher_counts[publisher_counts>95].plot(kind = "barh")
#plt.xticks(rotation = 45)
plt.title("Action türünde oyun çıkaran Publisher'lar - İlk 10", fontfamily = "serif", size = 12)
plt.show()
```

# En çok hangi türde oyun çıkarılmış?

```python
data["Genre"].value_counts()
```

```python
genres = data["Genre"].value_counts().sample(frac = 1)
genres.index.tolist()
```

```python
f, ax = plt.subplots(figsize = (8,4))
plt.bar(genres.index.tolist(), genres)
plt.xticks(rotation = 45)
plt.title("En çok hangi türde oyun çıkarılmış?", fontfamily = "serif", size = 12)
plt.show()
```

# Bölgelere göre Global satışlar

```python
global_sales_by_region = data.describe().columns[1:5].tolist()
global_sales_by_region
```

```python
na_sales = data["NA_Sales"].sum()
eu_sales = data["EU_Sales"].sum()
jp_sales = data["JP_Sales"].sum()
other_sales = data["Other_Sales"].sum()
```

```python
values = [na_sales, eu_sales, jp_sales, other_sales]
explode = [0.1, 0, 0, 0]
```

```python
f, ax = plt.subplots(figsize = (8,6))
plt.pie(values, labels = global_sales_by_region, explode = explode, startangle = 90, autopct = "%1.1f%%", shadow = True)
plt.title("Bölgelere göre Global satışlar %", fontfamily = "serif", size = 12)
plt.show()
```
